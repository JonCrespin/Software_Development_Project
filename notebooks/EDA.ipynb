```python
df= pd.read_csv('../vgchartz_2024.csv')
```

```python
df= df.drop(df.columns[0], axis =1)
df.info() 
df.head() 
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')
print(df['release_date'].head())
```

```python
df = df.dropna(subset=['release_date'])
df.info()
```

```python
df['last_update'] = pd.to_datetime(df['last_update'], format='%Y-%m-%d', errors='coerce')
print(df['last_update'].head())
```

```python
df = df.dropna(subset=['last_update'])
df.info()
```

```python
median_critic_score = df['critic_score'].median()
df['critic_score'] = df['critic_score'].fillna(median_critic_score)
print(df['critic_score'].median())
```

```python
df.info() 
df.head() 
```

```python
missing_developers = df[df['developer'].isna()]
print(missing_developers[['title', 'console']])
```

```python
df.loc[(df['title'] == 'My Friend Pedro') & (df['console'].isin(['NS', 'PC'])), 'developer'] = 'DeadToast Entertainment'
df.info()
```

```python
df.dropna(inplace= True) 
df.info()
```

```python
df['console'] = df['console'].replace({'XOne': 'XBOne','Series': 'Xbox Series','X360': 'XB360' })
start_year, end_year = 2000, 2024
console_choice = df[
    (df['release_date'].dt.year >= start_year) & 
    (df['release_date'].dt.year <= end_year)
]['console'].unique()
console_choice.sort()
console_choice
selected_menu = 'PS4'

df_filtered = df[(df.console == selected_menu) ]
```

```python
df_genre_developer_sales = df.groupby(['genre', 'developer'])['total_sales'].sum().reset_index()
df_genre_developer_sales
```

```python
sales_columns = ['na_sales', 'jp_sales', 'pal_sales', 'other_sales']
sales_columns
```

```python
def sales_vs_critic_score(df, include_genre=False):
    required_columns = ['critic_score', 'total_sales']
    if not all(col in df.columns for col in required_columns):
        raise ValueError("DataFrame must include 'critic_score' and 'total_sales' columns.")
    
    analysis_df = df[['critic_score', 'total_sales']].copy()
    if include_genre:
        analysis_df['genre'] = df['genre']
    
    analysis_df = analysis_df.dropna(subset=['critic_score', 'total_sales'])
    return analysis_df

include_genre = st.checkbox("Include Genre in Visualization")
if include_genre:
    analysis_df = sales_vs_critic_score(df_filtered, include_genre=True)
```

```python
analysis_df = sales_vs_critic_score(df_filtered, include_genre=False)

include_genre = st.checkbox("Include Genre in Visualization", key="include_genre_checkbox")
if include_genre:
    analysis_df = sales_vs_critic_score(df_filtered, include_genre=True)

fig2 = px.histogram(
    analysis_df,
    x='critic_score',
    y='total_sales',
    color='genre' if include_genre else None,
    nbins=20,
    title="Sales vs. Critic Score" + (" by Genre" if include_genre else ""),
    labels={'critic_score': 'Critic Score', 'total_sales': 'Total Sales'}
)
fig2.update_layout(bargap=0.1)

st.plotly_chart(fig2)
```
